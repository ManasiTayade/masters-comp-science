```python
x_train = traindata.drop(['Salary_range', 'Workclass', 'Education', 'Marital-status', 'Occupation', 'Relationship'], axis=1 )
x_train
```

```python
y_train = traindata['Salary_range']
y_train
```

```python
x_test = testdata.drop(['Salary_range', 'Workclass', 'Education', 'Marital-status', 'Occupation', 'Relationship'], axis=1 )
x_test
```

```python
y_test = testdata['Salary_range']
y_test
```

```python
labels = ['Age', 'Fnlwgt', 'Education-num', 'Race', 'Sex', 'Capital-gain', 'Capital-loss', 'Hours-per-week', 'Native_country']
labels
```

```python
estimator = DecisionTreeClassifier()
estimator.fit(x_train, y_train)
```

```python
graph = Source(tree.export_graphviz(estimator, out_file=None
   , feature_names=labels, class_names=['0', '1'] 
   , filled = True))
display(SVG(graph.pipe(format='svg')))
```

```python
y_pred = estimator.predict(x_test)
```

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
